```python
def getDataRate( h , noise=3.0673e-20):
  B=10**7 ; #Hz
  K=500;
  M=20;
  P=1; # Watt
  

  PSD = h * np.conj(h);
  rate = (B/(K+M-1)) * np.sum(np.log2(1+(P*PSD.real)/((B*noise))), axis = 0)
  rate=rate/1e6
  return rate
```

```python
import torch
def getDataRate_grad( ht_real,ht_imag , noise ):
  B=10**7  #Hz
  K=500
  M=20
  P=1 # Watt


  PSD = torch.mul(ht_real,ht_real)+torch.mul(ht_imag,ht_imag)

  rate = (B/(K+M-1)) * torch.sum(torch.log2(1+((P*PSD)/(B*noise))), axis = 0)
  rate=rate/1e6
  
  return rate


# # needed funtions
# theta=torch.from_numpy(given_best_config[0])
# theta_with_grad=torch.tensor(theta,dtype = torch.float32,requires_grad=True)


def get_grad(u,theta):


  ureal=u.real
  uimag=u.imag
  ut_real=torch.tensor(ureal,dtype = torch.float32)
  ut_imag=torch.tensor(uimag,dtype = torch.float32)
  # theta=torch.from_numpy(given_best_config[0])
  # theta_with_grad=torch.tensor(theta,dtype = torch.float32,requires_grad=True)
  # theta = 0.001 * np.random.random_sample((4096, 1)) - 1
  # theta_gpu = torch.tensor(theta, dtype = torch.float32,   requires_grad=True)
  h_real=torch.matmul(ut_real,theta)
  h_imag=torch.matmul(ut_imag,theta)
  rate=getDataRate_grad( h_real,h_imag , 3.0673e-20 )
  rate.backward()
  return np.array(theta.grad),rate 

  

def grad_opt(initial_conf  ,u,iteration):
  datarate_change=[]
  
  theta_with_grad=torch.tensor(initial_conf,dtype = torch.float32,requires_grad=True)
  for x in range(50):
    theta_without_grad=theta_with_grad.detach().numpy()
    # h=np.dot(u,theta_without_grad)
    # datarate=getDataRate_grad( h , 3.0673e-20 )
    # datarate_change.append(datarate)

    grad,rate =get_grad(u,theta_with_grad)
    datarate_change.append(rate.data.tolist())

    mask_pos=grad>0
    


    mask_neg=grad<0
    

    theta_without_grad[mask_pos]=1
    theta_without_grad[mask_neg]=-1

    theta_with_grad=torch.tensor(theta_without_grad,dtype = torch.float32,requires_grad=True)
    

  return theta_with_grad.detach().numpy() ,datarate_change,rate.data.tolist()

def grad_opt_final(u,user):
  lrate=[]
  lconfig=[]
  
  config,rates,best_rate=grad_opt(initial_conf=given_best_config[user] ,u=u,iteration=20)
  lrate.append(best_rate)
  lconfig.append(config)

  for x in tqdm(range(50)):
    random_conf = np.random.randint(0,2,(4096))
    random_conf=np.where(random_conf==0, -1, random_conf)
    config,rates,best_rate=grad_opt(initial_conf=random_conf ,u=u,iteration=20)
    lrate.append(best_rate)
    lconfig.append(config)
  return lrate,lconfig
```
